#2. 변수(variable)
변수(Variable)은 프로그램에서 값들이 저장되는 공간을 변수라고 합니다.
변수가 만들어지는 곳은 MAIN MEMORY에 만들어집니다.

```python
a = "Hello"
b = "World"

print(a+b)
```

변수(variable)란 데이터(data)를 저장하기 위해 프로그램에 의해 이름을 할당받는 메모리 공간을 의미합니다.

즉, 변수란 데이터를 저장하는 메모리의 공간입니다. 이렇게 저장된 값들은 변경을 할 수 있습니다.

​

1. 변수 명의 첫 문자는 언더 스코어 ( _ ) 또는 영문자로 시작되어야 합니다.

3. 변수 이름 사이에 공백을 포함하면 안 됩니다.

변수명에서 두 번째 문자부터는 알파벳, 숫자 그리고 밑줄 문자를 사용할 수 있습니다.

```python
# 리스트모양
li = [1,2,3,4,5]

print(li)
```

리스트 이름 = [요소값1, 요소값2, 요소값3]

리스트는 보다시피 이처럼 대괄호( [  ] )로 감싸져있다. 각 요솟값은 ","으로 구분해 준다.

```python
li1 = []
li2 = [1, 2, 3]
li3 = ["a", "b", "c"]
li4 = [1, 2, "a", "b"]
li5 = [1, 2, ["a", "b"]]
```

리스트 1번은 아무것도 포함 안된 빈 리스트이고

리스트 2번은 요솟값이 int 형이고

리스트 3번은 요솟값이 str 형이고

리스트 4번은 int형과 str형 두 개가 들어가 있고

리스트 5번에는 리스트 안에 리스트가 있습니다.

리스트_번호.png

다음처럼 리스트는 0번부터 시작한다.

```python
li1 = [1,2,3,4,5]
print(li1[0]) #1을 출력
print(li1[0] + li1[3]) #리스트0의 1과 리스트3의 4를 더한값 5 출력
```

리스트 더하기

```python
li1 = [1, 2, 3, 4]
li2 = [4, 5, 6, 7]
print(li1 + li2)
```

리스트 반복하기

```python
li1 = [8, 9, 10]
print(li1 * 4)
```

요소 추가(append)

```python
li1 = [1,2,3]
li1.append(15) #리스트 끝에 15를 추가
print(li1)
```

특정요소 삽입(insert)

```python
li1 = [1,2,3,4,5]
li1.insert(2,15) #2번자리에 15를 추가
print(li1)
```

특정위치 제거(del)

```python
li1 = [1,2]
del li1[0] #리스트 0번 제거
print(li1)
```

정렬(sort)

```python
li1 = [1, 3, 2, 4]
li1.sort()
print(li1)
```

리스트 뒤집기(reverse)

```python
li1 = [1, 2, 3, 4, 5]
li1.reverse()
print(li1)
```
